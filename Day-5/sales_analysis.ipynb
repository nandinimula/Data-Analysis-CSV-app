Below code creates a Sample-CSV Data for our Project.

```python
import pandas as pd

data = {
    "Date": [
        "2025-01-01", "2025-01-02", "2025-01-03", "2025-01-04", "2025-01-05",
        "2025-01-06", "2025-01-07", "2025-01-08", "2025-01-09", "2025-01-10"
    ],
    "Region": [
        "North", "South", "North", "East", "South",
        "East", "North", "South", "East", "North"
    ],
    "Product": [
        "Shampoo", "Soap", "Toothpaste", "Shampoo", "Shampoo",
        "Toothpaste", "Soap", "Toothpaste", "Soap", "Shampoo"
    ],
    "Sales": [120, 80, 150, 200, 90, 180, 100, 140, 160, 130]
}

df_sample = pd.DataFrame(data)
df_sample.to_csv("sales_data.csv", index=False)
print("✅ Sample sales_data.csv created")
```

The below cell loads Sample-CSV data and shows Preview.

```python
df = pd.read_csv("sales_data.csv")

print("First 5 rows of data:")
display(df.head())

print("\nData Summary:")
display(df.describe())
```

In the below cell we wrote code to arrange the sample data with the help of Group Sales by Region & Product.

```python
sales_by_region = df.groupby("Region")["Sales"].sum()
sales_by_product = df.groupby("Product")["Sales"].sum()

print("Total Sales by Region:")
display(sales_by_region)

print("\nTotal Sales by Product:")
display(sales_by_product)
```

This below cell code is used to Plot the Charts.

```python
import matplotlib.pyplot as plt

# Sales by Region
plt.figure(figsize=(6,4))
sales_by_region.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Sales by Region")
plt.ylabel("Sales")
plt.xlabel("Region")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Sales by Product
plt.figure(figsize=(6,4))
sales_by_product.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Total Sales by Product")
plt.ylabel("Sales")
plt.xlabel("Product")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
